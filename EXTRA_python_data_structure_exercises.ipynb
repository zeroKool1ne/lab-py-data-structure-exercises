### Exercise 3: Square Each Element
Given a list of integers, return a new list containing the squares of each number.

```python
numbers = [1, 2, 3, 4, 5]
running_sum_list = [0]            # I need a list here NOT just a variable
for n in numbers:                 # Create loop
  running_sum_list.append(n*n)    # Define what to be done in interation // append puts result of n*n into the new list. Without it, only 1 result

print(running_sum_list)
```

```python
squared = [n*n for n in numbers]    # SMARTER
print(squared)                      # Learning: you can put the mathematical equ in front of the loop definition
```

### Exercise 12: Count Character Frequency
Count the frequency of each character in a string and store it in a dictionary.

```python
input_string = "hello world"
dict_freq = {}

for char in input_string:
  dict_freq[char] = dict_freq.get(char, 0) +1

print(dict_freq)
```

### Exercise 13: Get Keys as List
